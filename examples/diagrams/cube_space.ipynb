```python
import networkx as nx
import matplotlib.pyplot as plt
import sys
import os
import numpy as np

# Add the project root to sys.path
project_root = os.path.abspath(os.path.join(os.getcwd(), "../.."))
sys.path.append(project_root)

import dagology as dag

np.random.seed(42)
```

```python
%matplotlib inline
```

```python
N = 6
D = 2
R = dag.cube_space_interval(N, D)
```

```python
G = dag.CubeSpaceGraph(lp=0.5).generate_graph(R, weighted=True)
G_TR = nx.transitive_reduction(G)
```

```python
pos = {i:G.nodes[i]['position']for i in G.nodes()}
labels = {i: str(i) for i in G.nodes()}
# labels = {i: f"{i}\n\n" + ", ".join(f"{t:.3f}" for t in pos[i]) for i in G.nodes()}
```

```python
plt.figure(figsize=(7,7))

nx.draw(G_TR, pos=pos, width=3)
nx.draw_networkx_edges(G, pos, width=1)
nx.draw_networkx_labels(G,pos,labels, font_size=10, font_color='white')
edge_labels = {i:f"{nx.get_edge_attributes(G,'weight')[i]: .3f}" for i in G_TR.edges}
nx.draw_networkx_edge_labels(G_TR, pos, edge_labels=edge_labels, font_color='blue')

plt.axis('off')
plt.savefig('../figures/cube/cube_greedy_example.svg')
plt.show()
```
